```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Фіксовані 3D-точки
A = np.array([1, 2, 3])
B = np.array([3, 1, 2])

fig = plt.figure(figsize=(18, 5))

# 1) Евклідова відстань
ax1 = fig.add_subplot(1, 3, 1, projection='3d')
ax1.scatter(*A, color='green', s=50)
ax1.scatter(*B, color='blue', s=50)
ax1.plot([A[0], B[0]], [A[1], B[1]], [A[2], B[2]], color='red', linewidth=2)
ax1.set_title('Евклідова відстань')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

# 2) Манхеттенська відстань (L1-путь)
ax2 = fig.add_subplot(1, 3, 2, projection='3d')
ax2.scatter(*A, color='green', s=50)
ax2.scatter(*B, color='blue', s=50)
# Ломаная через проміжні точки
M1 = np.array([B[0], A[1], A[2]])
M2 = np.array([B[0], B[1], A[2]])
ax2.plot([A[0], M1[0]], [A[1], M1[1]], [A[2], M1[2]], color='red', linewidth=2)
ax2.plot([M1[0], M2[0]], [M1[1], M2[1]], [M1[2], M2[2]], color='red', linewidth=2)
ax2.plot([M2[0], B[0]], [M2[1], B[1]], [M2[2], B[2]], color='red', linewidth=2)
ax2.set_title('Манхеттенська відстань')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')

# 3) Косинусна схожість
ax3 = fig.add_subplot(1, 3, 3, projection='3d')
# Нормалізовані вектори
a = A / np.linalg.norm(A)
b = B / np.linalg.norm(B)
# Відображення векторів
ax3.plot([0, a[0]], [0, a[1]], [0, a[2]], color='green', linewidth=2)
ax3.scatter(*a, color='green', s=50)
ax3.plot([0, b[0]], [0, b[1]], [0, b[2]], color='blue', linewidth=2)
ax3.scatter(*b, color='blue', s=50)
# Проекція a на b
cos_theta = np.dot(a, b)
a_proj = cos_theta * b
ax3.plot([0, a_proj[0]], [0, a_proj[1]], [0, a_proj[2]], color='red', linewidth=2)
# Креслення дуги кута
theta = np.arccos(cos_theta)
v = b - cos_theta * a
v /= np.linalg.norm(v)
t = np.linspace(0, theta, 50)
r = 0.5  # радіус дуги
arc = np.outer(np.cos(t), a) + np.outer(np.sin(t), v)
arc *= r
ax3.plot(arc[:,0], arc[:,1], arc[:,2], color='red')
# Підписи
mid = len(t) // 2
ax3.text(arc[mid,0], arc[mid,1], arc[mid,2], r'$\theta$', color='red')
ax3.text(a_proj[0], a_proj[1], a_proj[2], r'$\cos(\theta)$', color='red')
ax3.set_title('Косинусна схожість')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_zlabel('Z')

# Встановлюємо однакові межі для всіх осей
lims = np.array([A, B, [0,0,0]]).flatten()
min_lim, max_lim = lims.min() - 0.5, lims.max() + 0.5
for ax in (ax1, ax2, ax3):
    ax.set_xlim(min_lim, max_lim)
    ax.set_ylim(min_lim, max_lim)
    ax.set_zlim(min_lim, max_lim)

plt.tight_layout()
plt.show()
```
